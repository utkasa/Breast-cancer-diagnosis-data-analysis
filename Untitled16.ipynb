Data Cleaning

```python
df.isnull().sum()
```

```python
df["node-caps"]=df["node-caps"].replace("no', False")
df["node-caps"]=df["node-caps"].replace("yes', True")
```

```python
df["irradiate"]=df["irradiate"].replace("no', False")
df["irradiate"]=df["irradiate"].replace("yes', True")
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['deg-malig'] = df['deg-malig'].str.replace('"','').str.replace('','').str.replace("'",'').astype(int)
```

```python
df['inv-nodes']=df['inv-nodes'].str.replace("'",'')
```

```python
df['tumer-size']=df['tumer-size'].str.replace('"','').str.replace("'",'')
```

```python
df['menopause']=df['menopause'].str.replace("'","")
df['breast'] = df['breast'].str.replace("'","")
df['breast-quad'] = df['breast-quad'].str.replace("'","")
df['age'] = df['age'].str.replace("'","")
```

```python
df.rename(columns={'class':'recurrence'}, inplace=True)
```

```python
df['recurrence']=df['recurrence'].replace("recurrence-events'",True)
df['recurrence']=df['recurrence'].replace("no-recurrence-events'",False)
```

```python
df["node-caps"]=df["node-caps"].replace("no'", False)
df["node-caps"]=df["node-caps"].replace("yes'", True)
df["irradiate"]=df["irradiate"].replace("no'", False)
df["irradiate"]=df["irradiate"].replace("yes'", True)
```
